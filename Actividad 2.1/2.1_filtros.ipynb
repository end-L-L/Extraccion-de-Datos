```python
import pandas as pd
```

```python
# Carga desde un archivo .xlsx sin índice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
# Filtro 1
# Filtrar por la columna NOMBRE_CLIENTE con valor "CPEG SOLUTIONS"
filtro1=df[(df["NOMBRE_CLIENTE"] == "CPEG SOLUTIONS")]
filtro1
```

```python
# Filtro 2
# Filtrar por índices de filas no consecutivas (700, 706 y 709)
filtro2 = df.iloc[[700,706,709], :]
filtro2
```

```python
# Filtro 3
# Filtrar por índices de columnas consecutivas (0, 1 y 2)
filtro3 = df.iloc[ : , 0:3]
filtro3
```

```python
# Filtro 4
# Paso 1: Cargar el archivo estableciendo la columna "NOMBRE_CLIENTE" como índice
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
df1
```

```python
# Paso 2: Seleccionar valores específicos de las filas "REVUELTA MONTECARLO" y "CPEG SOLUTIONS" 
# para la columna "NOMBRE_VENDEDOR"
filtro4 = df1.loc[["REVUELTA MONTECARLO", "CPEG SOLUTIONS"], ["NOMBRE_VENDEDOR"]]
filtro4
```

```python
# Filtro 5
# Mostrar las primeras 5 filas (cabecera)
filtro5 = df.head()
filtro5
```

```python
# Filtro 6
# Mostrar las últimas 5 filas (cola)
filtro6 = df.tail()
filtro6
```

```python
# Filtro 7
# Filtrar donde el valor de la columna "PRECIO_UNITARIO" sea mayor a 25000
filtro7 = df[df["PRECIO_UNITARIO"] > 25000]
filtro7
```

```python
# Filtro 8
# Filtrar donde "PRECIO_UNITARIO" sea mayor a 25000 y "COSTO_TOTAL_CALCULADO" sea menor a 100000
filtro8 = df[(df["PRECIO_UNITARIO"] > 25000) & (df["COSTO_TOTAL_CALCULADO"] < 100000)]
filtro8
```

```python
# Filtro 9
# Filtrar donde "PRECIO_UNITARIO" sea mayor a 25000 o "COSTO_TOTAL_CALCULADO" sea mayor a 100000
filtro9 = df[(df["PRECIO_UNITARIO"] > 25000) | (df["COSTO_TOTAL_CALCULADO"] > 100000)]
filtro9
```

```python
# Filtro 10
# Filtrar con operador "not" (~) donde el valor de la columna "FECHA_DOC" no es mayor a "2022-01-05"
filtro10 = df[~(df["FECHA_DOC"] > "2022-01-05")] 
filtro10
```
