```python
import pandas as pd
```

```python
#df = pd.read_csv('Florencia.csv')
df = pd.read_csv('Florencia.csv', index_col=0)
df
```

```python
df.info()
```

```python
# Filtro A
# a) los registros de 5 "host name" diferentes

# obtener una lista de 5 nombres √∫nicos de "host_name"
nombres_unicos = df['host_name'].dropna().unique()[:5]  # Seleccionamos los primeros 5 √∫nicos

# filtrar el DataFrame para incluir solo registros de esos nombres
filtro1 = df[df['host_name'].isin(nombres_unicos)]
filtro1.to_csv('filtro1.csv')
filtro1
```

```python
# Filtro B
# b) Los registros de los hosts que se hallan unido a Airbnb "antes del a√±o 2019".

filtro2 = df[df['host_since'] < '2019-01-01']
filtro2.to_csv('filtro2.csv')
filtro2
```

```python
# Filtro C
# c) Los registros de los hosts que responden "a m√°s tardar en 1 d√≠a".
filtro3 = df[df['host_response_time'] == 'within a day']
filtro3.to_csv('filtro3.csv')
filtro3
```

```python
# Filtro D
# d) Los registros que tengan "al menos 5 accommodates".
filtro4 = df[df['accommodates'] >= 5]
filtro4.to_csv('filtro4.csv')
filtro4
```

```python
# Filtro E
# e) Los registros de los tipos de cuarto: "Entire home" y "Private room".
filtro5 = df[df['room_type'].isin(['Entire home/apt', 'Private room'])]
filtro5.to_csv('filtro5.csv')
filtro5
```

```python
# Filtro F
# f) Los registros que "a lo mucho cuenten con 4 camas".
filtro6 = df[df['beds'] <= 4]   
filtro6.to_csv('filtro6.csv')
filtro6
```

```python
# Filtro G
# g) Los registros que "tengan 1 ba√±o o que sean superhost" pero con la condici√≥n de que "todos sean instant_bookable".
filtro7 = df[((df['bathrooms'] == 1) | (df['host_is_superhost'] == 't')) & (df['instant_bookable'] == 't')]
filtro7.to_csv('filtro7.csv')
filtro7
```

```python
# Filtro H
# h) Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3".
filtro8 = df[(df['review_scores_cleanliness'] < 3) & (df['review_scores_rating'] > 3)]
filtro8.to_csv('filtro8.csv')
filtro8
```
